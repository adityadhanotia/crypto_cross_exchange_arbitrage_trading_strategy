an attempt to create a uniform fetch function 

```python
symbol = 'MATIC/USDT'

def fetch_recent_trades(exchange, symbol, since):
    params = {}
    if exchange.id == 'coinbase':
        until = exchange.milliseconds()
        params['until'] = until
    
    trades = exchange.fetch_trades(symbol, since=since, params=params)
    return pd.DataFrame(trades)
```

```python
#calculate the mean buy/sell ratios 

def calculate_mean_buy_sell_ratio(trades, interval_hours=4):
    # Convert timestamp to datetime and set as index
    trades['date_time'] = pd.to_datetime(trades['timestamp'], unit='ms')
    trades.set_index('date_time', inplace=True)

    # Resample trades into 1-minute intervals
    resampled = trades.resample(f'{interval_hours}H')

    ratios = []
    for interval_start, interval_trades in resampled:
        buy_volume = interval_trades[interval_trades['side'] == 'buy']['amount'].sum()
        sell_volume = interval_trades[interval_trades['side'] == 'sell']['amount'].sum()
        
        if sell_volume != 0:
            ratio = buy_volume / sell_volume
        else:
            ratio = 1  # come back to this, should it be 0? 
        
        ratios.append(ratio)
        # print(f"Interval starting at {interval_start}: ratio = {ratio:.2f}")

    # Calculate the mean of the ratios
    mean_ratio = sum(ratios) / len(ratios) if ratios else 1 # come back to this, should it be 0?

    return mean_ratio
```

```python
def make_decision(kraken_mean_ratio, coinbase_mean_ratio):
    if kraken_mean_ratio > coinbase_mean_ratio:
        return 'sell_kraken_buy_coinbase'
    elif coinbase_mean_ratio > kraken_mean_ratio:
        return 'sell_coinbase_buy_kraken'
    return 'no_trade'
```

```python

# Test the function
decision = make_decision(kraken_mean_ratio, coinbase_mean_ratio)
print(decision)
```

```python

def execute_trade(decision, amount = 1):
    if decision == 'sell_kraken_buy_coinbase':
        return f"Selling {amount} MATIC-USDT on Kraken\nBuying {amount} MATIC-USDT on Coinbase" 
    elif decision == 'sell_coinbase_buy_kraken':
        return f"Selling {amount} MATIC-USDT on Coinbase\nBuying {amount} MATIC-USDT on Kraken"
    else:
        return "No trade executed"

```

## Automated Trading ##

```python
def main():
    start_time = datetime.utcnow()
    end_time = start_time + timedelta(minutes=5) # run for __ minutes 
    iteration_count = 0
    trade_count = 0 


    while datetime.utcnow() < end_time:
        try:
            iteration_count += 1
            current_time = datetime.utcnow()
            print(f"\nIteration {iteration_count} at {current_time}")

            since = int((current_time - timedelta(hours=24)).timestamp() * 1000)

            # Fetch recent trades
            kraken_trades = fetch_recent_trades(kraken, symbol, since)
            coinbase_trades = fetch_recent_trades(coinbase, symbol, since)

            # Calculate mean buy/sell ratios
            kraken_mean_ratio = calculate_mean_buy_sell_ratio(kraken_trades)
            coinbase_mean_ratio = calculate_mean_buy_sell_ratio(coinbase_trades)

            print(f"Kraken mean buy/sell ratio: {kraken_mean_ratio:.4f}")
            print(f"Coinbase mean buy/sell ratio: {coinbase_mean_ratio:.4f}")

            # Make decision and execute trade
            decision = make_decision(kraken_mean_ratio, coinbase_mean_ratio)
            if decision != 'no_trade':
                trade_count += 1
            print(execute_trade(decision))

            # Calculate and print time remaining
            time_remaining = end_time - current_time
            print(f"Time remaining: {time_remaining}")

            # Wait for 1 minute before next iteration
            time.sleep(60)

        except Exception as e:
            print(f"An error occurred: {e}")
            time.sleep(60)  # Wait a minute before retrying

    print(f"Strategy execution completed. Total iterations: {iteration_count}, Total trades executed: {trade_count}")

# Run the main function
main()
```
